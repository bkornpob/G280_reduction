```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10,10))

x = '/Users/kbhirombhakdi/_work/_G280_reduction/170822/G102/g102_spec.dat'
x = pd.read_csv(x,sep=' ',comment='#',index_col=False
                ,names=['wavelength','count_rate','sensitivity','flam','fnu'])
m = np.argwhere(x['wavelength']>8000.).flatten()
m = np.intersect1d(m,np.argwhere(x['wavelength']<11400.).flatten())
plt.plot(x['wavelength'][m],x['flam'][m])

x = '/Users/kbhirombhakdi/_work/_G280_reduction/170822/G141/g141_spec.dat'
x = pd.read_csv(x,sep=' ',comment='#',index_col=False
                ,names=['wavelength','count_rate','sensitivity','flam','fnu'])
m = np.argwhere(x['wavelength']>11000.).flatten()
m = np.intersect1d(m,np.argwhere(x['wavelength']<16000.).flatten())
plt.plot(x['wavelength'][m],x['flam'][m])

plt.yscale('log')
plt.xscale('log')
plt.xlabel('Wavelength ($\AA$)')
plt.ylabel('erg/s/cm$^2$/$\AA$')
```
